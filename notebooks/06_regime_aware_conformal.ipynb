# 06: Regime-Aware Conformal Prediction

## Goals
1. Combine HMM regime detection with conformal prediction
2. Maintain separate nonconformity score distributions per regime
3. At each time step: detect regime via forward pass, select corresponding score distribution, build interval
4. Compare three methods: static conformal, ACI, and regime-aware conformal
5. Show regime-aware conformal achieves better per-regime coverage than ACI, especially at regime transitions

## Key Idea
ACI adapts intervals reactively — it waits for misses to accumulate before widening. 
Regime-aware conformal adapts proactively — it detects the regime shift and immediately 
switches to the appropriate score distribution, avoiding the lag.

## Setup
- Same synthetic non-stationary data as notebook 05
- HMM trained on early portion of data to learn regime parameters
- Separate calibration score sets for calm and crisis regimes
- α = 0.1 (target 90% coverage)
